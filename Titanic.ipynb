```python
import pandas as pd
location = 'titanic3.xls'
df = pd.read_excel(location)
df.head()
```

```python
##hasfamily: 0 or 1
df.loc[df['sibsp']+df['parch'] < 1,'hasfamily'] = 0
df.loc[df['sibsp']+df['parch'] > 0,'hasfamily'] = 1
df.head()
```

```python
#status: 0 lost at sea or 1 body recovered or 2 survived
df.loc[df['body'].isnull() & df['boat'].isnull(),'status'] = 0
df.loc[df['body'].notnull() ,'status'] = 1
df.loc[df['survived'] == 1 ,'status'] = 2
df.head()
```

```python
##accomodation: 0 for steerage or 1 for cabin
df.loc[df['cabin'].isnull(), 'accomodation'] = 0
df.loc[df['cabin'].notnull(), 'accomodation'] = 1
df.head()
```

```python
noboat = df['boat'].isnull()
```

```python
##Then remove the rows that are truly missing data.

df[df['name'].isnull()].dropna()
df[df['age'].isnull()].dropna()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ fare + hasfamily -1', data=df).fit()
result.summary()
```
